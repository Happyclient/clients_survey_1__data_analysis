<a id="cleaning"></a>
<h2>DATA CLEANING PHASE</h2>

```python
# check for duplicated rows
df.duplicated().sum()
```

```python
# check for and handle null values
df.isnull().sum()
```

```python
# create index column - id
df['id'] = range(1, len(df) + 1)
df = df.set_index('id')
```

```python
# drop unecessary columns
df = df.drop('Timestamp', axis=1)
df
```

```python
# give columns appropriate names
new_column_names = [
    "gender",
    "age_range",
    "marital_status",
    "education_level",
    "employment_status",
    "main_telecom",
    "social_channels",
    "challenges_in_support",
    "best_social_for_support",
    "comfort_with_chatbot",
    "csat_with_main_telecom_support",
    "ug_support_rating",
    "website_feat",
    "mobile_app_feat",
    "text_chat_feat",
    "video_and_screen_share_feat",
    "local_languages_feat",
    "send_images_feat",
    "searchable_self_help_feat",
    "recommended_solutions_feat",
    "issue_progress_status_feat",
    "schedule_support_feat",
    "rate_and_review_agent_feat",
    "discussion_forum_feat",
    "important_factor_in_support",
    "necessity_of_app_for_support"
]
df = df.set_axis(new_column_names, axis=1, inplace=False)
```

```python
# clean column values

# clean gender column
df = df.replace({'Male': 'male', 'Female': 'female'})

# clean age column
df = df.replace({'below 19 years': '<19', '20 years - 35 years': '20-35', '35 years - 45 years': '35-40', 'above 45 years': '>45' })

# clean marital status column
df = df.replace({'Cohabiting': 'single', 'Single': 'single', 'Married': 'married'})

# clean education level column
df = df.replace({'Undergraduate': 'undergraduate', 'Secondary school': 'secondary', 'Vocational college': 'vocational', 'Post graduate and above': 'postgraduate'})

# clean employment status column 
df = df.replace({'Student': 'student', 'Self employed': 'self employed', "Unemployed, looking for work": 'unemployed', 'Employed on a salary': 'salary earner', 'Retired': 'retired', 'Volunteering ': 'unemployed', 'Student & self employed ': 'self employed', 'Confidential': 'unknown_status'})

# clean main telecom column
df = df.replace({'MTN': 'Mtn', 'AIRTEL': 'Airtel', 'Lyca mobile': 'Lycamobile', 'Lyca Mobile': 'Lycamobile'})

# clean social channels column
unclean_socials = list(df['social_channels'])
clean_socials = []

for social_group in list(df['social_channels']):
    clean_socials.append(social_group.lower().replace(';', ' | '))
    
df = df.replace(unclean_socials, clean_socials)

# clean challenges in support column
unclean_challenges = list(df['challenges_in_support'])
clean_challenges = []

for challenges_group in list(df['challenges_in_support']):
    clean_challenges = challenges_group.lower().replace(';', ' | ')
    clean_challenges = clean_challenges.replace('many steps have to be followed to talk to a customer care agent.', 'many steps to talk to an agent')
    clean_challenges = clean_challenges.replace('no/late responses to my customer support complaints on social media.', 'no or late responses to tickets')
    clean_challenges = clean_challenges.replace('connected to an agent with low technical knowledge needed to handle my issue.', 'connected to agent with low technical knowledge')
    clean_challenges = clean_challenges.replace('no/low availability of customer support agents late in the night.', 'no or few agents availabe at night')
    clean_challenges = clean_challenges.replace('no/low availability of customer support agents over the weekends.', 'no or few agents availabe on weekends')
    unclean_challenges.append(clean_challenges)
    
# clean best social for support column
df = df.replace({'Calls': 'call', 'Phone call': 'call', 'call': 'call', 'Messages': 'sms', 'Feedback message ': 'sms', 'Facebook': 'facebook', 'Phone call ': 'call'})

# clean feat columns
df = df.replace({'I am not sure': 'not sure'})
df['website_feat'].unique()

# clean important factor in support column
df = df.replace({'Speed at which my issue is sorted.': 'speed', 'How easy it is for me to have my issue sorted.': 'ease', 'Ability to have my issue sorted on my own (without having to talk to an agent or someone)': 'self service'})

# clean necessity of app for support column
df = df.replace({'Very necessary': 'very necessary', 'Not necessary at all': 'not necessary', 'not sure': 'not sure'})

df
```

<a id="exploration"></a>
<h2>DATA EXPLORATION PHASE</h2>

<h3>Reusable Functions:</h3>

```python
# reusables

# colors
base_color = '#273c75'
target_color = '#44bd32'

# Dataframe Subsets
feature_columns = ["website_feat", "mobile_app_feat", "text_chat_feat", "video_and_screen_share_feat", "local_languages_feat", "send_images_feat", "searchable_self_help_feat", "recommended_solutions_feat", "issue_progress_status_feat", "schedule_support_feat", "rate_and_review_agent_feat", "discussion_forum_feat"]

features_df = df[feature_columns]
```

<a id="1d-1"></a><h3>1. Which <u>telecom service provider</u> is considered the primary service provider by most respondents?</h3>

```python
main_telecom_count = pd.Series(df['main_telecom']).value_counts(ascending=False)

# plot a pie
main_telecom_count.plot(kind='bar', figsize=(10,5), fontsize=12, color=[target_color, base_color, base_color])
plt.title('The most primary telecom service provider to respondents')
plt.xlabel('Telecom Companies')
plt.ylabel('Primary Subscribers')
plt.xticks(rotation=0, horizontalalignment='center')
```

<a id="1d-2"></a><h3>2. On which <u>social media platform</u> do most respondents have an account?</h3>

```python
total_social_channels_instances = df['social_channels'].str.cat(sep=' | ')
socials_channels_count = pd.Series(total_social_channels_instances.split(' | ')).value_counts(ascending=False)

# plot a pie
socials_channels_count.plot(kind='bar', figsize=(10,5), fontsize=12, color=[target_color, base_color, base_color, base_color, base_color])
plt.title('Most common social media channel among respondents')
plt.xlabel('Social Media Channels')
plt.ylabel('Social Media Channel Users')
plt.xticks(rotation=0, horizontalalignment='center')
```

<a id="1d-3"></a><h3>3. Which <u>two social media channels</u> are considered the best for sending a complaint to a telecom company?</h3>

```python
main_telecom_count = pd.Series(df['best_social_for_support']).value_counts(ascending=False)

# plot a pie
main_telecom_count.plot(kind='bar', figsize=(10,5), fontsize=12, color=[target_color, target_color, base_color, base_color, base_color, base_color, base_color])
plt.title('The two best social media channel for receiving customer support among respondents')
plt.xlabel('Social media channels')
plt.ylabel('Users')
plt.xticks(rotation=0, horizontalalignment='center')
```

<a id="1d-4"></a><h3>4. Which <u>three features</u> are considered most essential by customers in a customer support platform?</h3>

```python
feat_columns_labels = [
    "Website",
    "Mobile app",
    "Text chat",
    "Video/Screen sharing",
    "Local languages",
    "Share images",
    "Self help library",
    "Recommend solutions",
    "View issues status",
    "Schedule support",
    "Rate/Review agents",
    "Discussion forum"
]

essentials_counts_per_column = (features_df == 'essential').sum().to_numpy()

essential_counts_df = pd.DataFrame(data=essentials_counts_per_column, columns=['essential_counts'], index=feat_columns_labels).sort_values(by=['essential_counts'], ascending=False)
essential_counts_df.plot.bar(figsize=(12,5), fontsize=12, color=[target_color, base_color])

plt.title('The three most essential features to respondents')
plt.xlabel('Features')
plt.ylabel('Users that find it essential')
plt.xticks(rotation=80, horizontalalignment='center')
```

<a id="1d-5"></a><h3>5. Which <u>factor</u> is most important to customers when contacting customer support?</h3>

```python
important_factors_count = pd.Series(df['important_factor_in_support']).value_counts(ascending=False)

# plot a bar graph
important_factors_count.plot(kind='bar', figsize=(10,5), fontsize=12, color=[target_color, base_color, base_color])
plt.title('The most important factor when contacting customer support among respondents')
plt.xlabel('Factors')
plt.ylabel('Users who find it most important')
plt.xticks(rotation=0, horizontalalignment='center')
```

<a id="1d-6"></a><h3>6. How <u>comfortable</u> are customers with texting a chatbot when receiving customer support?</h3>

<a id="1d-7"></a><h3>7. How <u>necessary</u> is a platfrom dedicated to creating and managing customer support complaints to respondents?</h3>

```python
necessity_of_support_platform_counts = pd.Series(df['necessity_of_app_for_support']).value_counts(ascending=False)

necessity_of_support_platform_counts.plot.bar(figsize=(12,5), fontsize=12, color=[target_color, base_color, base_color])

plt.title('How necessary respondents find a customer support platform')
plt.xlabel('Necessity')
plt.ylabel('Number of users')
plt.xticks(rotation=0, horizontalalignment='center')
```
